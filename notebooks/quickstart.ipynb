## 1. Generate a motion & view the canonical skeleton
RobotDance standardizes a **Motion Intermediate Representation (RD-MIR)** — a canonical 19-joint skeleton. Here we synthesize a dance and render it (a real video would land in the same RD-MIR via the `perception` extra).

````python
!python -m robotdance_core.cli synth -o dance.json --duration 4
!python -m robotdance_core.cli view dance.json -o skeleton.gif --stride 2
display(Image(filename="skeleton.gif"))
````

## 2. Same motion → many humanoids  🤖🤖🤖
The headline feature: **one motion, retargeted onto six different bodies**. Each robot has different limb proportions, so the retarget normalizes morphology while preserving bone directions (`height_scale` differs per robot, foot-sliding stays tiny).

````python
!python -m robotdance_core.cli demo-multi -o multi.gif --stride 2 \
    --robots unitree_g1 unitree_h1 unitree_h2 booster_t1 apptronik_apollo fourier_n1
display(Image(filename="multi.gif"))
````

## 3. Retarget quality metrics
Retargeting reports geometric fidelity — `bone_direction_cosine` (≈1.0 = directions preserved), `foot_sliding`, and joint-ROM violations. For physical feasibility (torque / balance / fall) use the `sim` extra + `validate-sim` (MuJoCo).

````python
!python -m robotdance_core.cli retarget dance.json -o g1.rdmotion.json --robot unitree_g1
````

## Next: your own video → humanoid
This tour used a synthetic motion to stay light. The full **"Shorts to humanoid"** path takes a real monocular video → MediaPipe Pose → RD-MIR → retarget. Install the perception extra and point `extract` at a clip:

```bash
pip install "robotdance[perception] @ git+https://github.com/rsasaki0109/HumanoidBattle.git"
python -m robotdance_core.cli extract your_clip.mp4 -o clip.rdmir.json
python -m robotdance_core.cli retarget clip.rdmir.json --robot unitree_g1 -o robot.rdmotion.json
```

⚠️ Only process video you have the rights to. RobotDance is license-safe by design — it never redistributes raw video/mocap, only numeric motion and renders.
